```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df_t=pd.read_excel('/content/Telco - classification or clustering.xls')
df_t.head()
```

```python
X=df_t.drop('churn',axis=1)
Y=df_t['churn']
```

```python
X=X.fillna(0)
```

```python
#نرمال سازی یک روش به منظور هم اسکیل کردن تمام فیچر ها
# در روش مین مکس داده ها به فضای صفر تا یک می روند
#روش دیگر استاندارد سازی است که  دادهای کمتر از میانگین را از صفر کوچکتر می کرد و بیشتر ها را از صفر بزرگتر 
from sklearn.preprocessing import MinMaxScaler
MMS=MinMaxScaler()
MMS.fit(X)
Xnormal=MMS.transform(X)
#Xnomal=MMS.fit_transform(X)
Xnormal=pd.DataFrame(Xnormal , columns=X.columns)# برای اینکه اسم سر ستون ها به حالت خود باقی بماند و به عدد تبدیل نشود این دستور کالمن را می نویسیم.
Xnormal
```
