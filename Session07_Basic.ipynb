```python
import pandas as pd
import numpy as np
```

```python
#ساخت دیتا فریم
df1=pd.DataFrame({'A':[1,2,3,4],'B':[1,2,3,4],'C':[2,3,4,5]})
df2=pd.DataFrame({'A':[3,4,5,6],'B':[4,5,6,7],'D':[6,7,8,9]})
df1
```

```python
# سورت ستون ها در دستور اپند
DF=df2.append(df1,sort=True)
DF
```

```python
# اپند چند دیتا فریم
DF=df2.append([df1,df2],sort=True)
DF
```

```python
# مشکل این اپند بودن ایندکس تکراری است
df_=DF.reset_index()
df_
```

```python
#حذف ستون ایندکس
df_=DF.reset_index().iloc[:,1:]
df_
```

```python
# دستور جوین یا مرج
df_j=pd.merge(df1,df2,how='inner',on="A")#inner,right,outer
df_j
```

```python
df_j=pd.merge(df1,df2,how='left',on='A')
df_j
```

```python
df_j=pd.merge(df1,df2,how='left',on=['A','B'])
df_j
```
